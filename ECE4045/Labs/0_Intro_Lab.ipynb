#### Statistical Tools  
**Example — Exponential Distribution (Step-by-Step Explanation)**



#### Import numerical and plotting libraries

```python
import numpy as np                     # NumPy for numerical operations, arrays, and statistics
import matplotlib.pyplot as plt       # pyplot module for plotting graphs
```

#### Generate random samples from an exponential distribution

- `scale = 1` → mean = 1 (rate λ = 1 for exponential distribution)  
- `size = 1000` → generates 1000 random samples  

```python
samples = np.random.exponential(scale=1, size=1000)
```



#### Compute basic statistics

```python
print('Mean:', np.mean(samples))      # Average value of generated samples
print('Variance:', np.var(samples))   # Spread of the data (variability)
```

#### Plot histogram to visualise the distribution

```python
plt.hist(samples, bins=50)            # Create histogram with 50 bins
plt.title('Exponential Distribution') # Add title to plot
plt.show()                            # Display the figure
```

#### What each step does

- `import numpy as np` → loads numerical tools for arrays, random numbers, and statistics.  
- `import matplotlib.pyplot as plt` → loads plotting functions for visualising data.  
- `np.random.exponential(...)` → generates random values representing exponential behaviour (e.g., inter-arrival times in network traffic).  
- `np.mean()` and `np.var()` → compute statistical properties of the generated data.  
- `plt.hist()` → visualises the distribution shape.  
- `plt.show()` → renders the plot.
