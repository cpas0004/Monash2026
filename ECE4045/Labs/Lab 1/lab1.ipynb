```python
import numpy as np
import matplotlib.pyplot as plt
# Set the parameters for the exponential distribution
scale_parameter = float(input("Enter the scale parameter lambda for the exponential distribution: "))
# Input number of exponential random variables to generate
num_samples = int(input("Enter the number of random variables in the exponential distribution: "))
# Generate exponential random variables
exponential_rvs = np.random.exponential(scale=scale_parameter, size=num_samples)
# Calculate mean and variance
mean = np.mean(exponential_rvs)
variance = np.var(exponential_rvs)
print("Mean:", mean)
print("Variance:", variance)
# Plot the histogram of the generated exponential random variables
plt.hist(exponential_rvs, bins=30, density=True, alpha=0.6, color='g', edgecolor='b')
# Plot the theoretical exponential distribution
x = np.linspace(0, 10, num_samples)
plt.plot(x, np.exp(-x), 'r-', lw=2, label='Exponential PDF')
plt.title('Histogram of Exponential Random Variables')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```
