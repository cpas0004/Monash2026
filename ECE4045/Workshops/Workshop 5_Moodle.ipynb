## Task 1-2 Instruction
Generate exponential interarrival times and plot the event times 

```python
# TASK 1-2
import simpy
import numpy as np
import matplotlib.pyplot as plt


inter_arrivals = []
event_times = []   # actual DES event times

def my_process(env):
    lam = 10 # rate parameter for the exponential distribution
    value = 42

    # Repeat the event-generation process 10 times
    for _ in range(10):
        t1 = env.now
        Inter = np.random.exponential(1/lam)  # Generate inter-arrival time
        yield env.timeout(Inter)  # Wait for the inter-arrival time
        t2 = env.now
        time_diff = t2 - t1
        inter_arrivals.append(time_diff)
        event_times.append(t2)
        print(f"Inter arrival time = {time_diff}, generated inter = {Inter}, value = {value}")
    
        # Increase the value after each event
        value+=1

env = simpy.Environment()
env.process(my_process(env))
env.run( )   # run simulation until time = 5

plt.figure()
plt.vlines(event_times, 0, 1, linestyles='dashed')
plt.plot(event_times, np.ones(len(event_times)), 'o')
plt.xlabel("Event number")
plt.ylabel("Inter-arrival time")
plt.title("Inter-arrival Times")
plt.grid(True)
plt.show()



    
```

```python
"""
Event Latency example

Covers:

- Resources: Store

Scenario:
  This example shows how to separate the time Avg_del of events between
  processes from the processes themselves.

When Useful:
  When modeling physical things such as cables, RF propagation, etc.  it
  better encapsulation to keep this propagation mechanism outside of the
  sending and receiving processes.

  Can also be used to interconnect processes sending messages

Example by:
  Keith Smith
  https://simpy.readthedocs.io/en/latest/examples/latency.html
"""

import simpy
import numpy as np
SIM_DURATION = 100

class Cable:
    """This class represents the propagation through a cable."""

    def __init__(self, env, Avg_del):
        self.env = env
        self.Avg_del = Avg_del
        self.store = simpy.Store(env)
        
    def latency(self, value):
        # Draw propagation delay from exponential distribution with mean self.Avg_del
        delay = np.random.exponential(self.Avg_del)
        yield self.env.timeout(delay)
        self.store.put(value)

    def put(self, value):
        self.env.process(self.latency(value))

    def get(self):
        return self.store.get()

def sender(env, cable, pktgen_time):
    """A process which randomly generates messages."""
    while True:
        # wait for next transmission (exponential inter-arrival time)
        inter_arrival = np.random.exponential(pktgen_time)
        yield env.timeout(inter_arrival)
        cable.put(f'Sender sent this at {env.now}')

def receiver(env, cable):
    """A process which consumes messages."""
    while True:
        # Get event for message pipe
        msg = yield cable.get()
        print(f'Received this at {env.now} while {msg}')

# Setup and start the simulation
print('Event Latency')
env = simpy.Environment()

cable = Cable(env, 5)
env.process(sender(env, cable, 10))
env.process(receiver(env, cable))

env.run(until=SIM_DURATION)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
def inter_arrival_times(rate, time_limit):
    """
    Simulate a Poisson process given a rate (λ) and time limit.

    Parameters:
    rate (float): The rate parameter λ of the exponential distribution.
    time_limit (float): The total time duration to simulate.

    Returns:
    event_times (list): The times at which events occur (t_1, t_2, ... ≤ time_limit).
    interarrival_times (list): The interarrival times actually used (Δ_1, Δ_2, ...).
    """
    event_times = []
    interarrival_times = []
    current_time = 0.0

    while True:
        inter = np.random.exponential(1.0 / rate)
        if current_time + inter > time_limit:
            break  # do not include the overshoot interval
        current_time += inter
        event_times.append(current_time)
        interarrival_times.append(inter)

    return event_times, interarrival_times
def simulate_multiple_processes(rate, time_limit, num_simulations):
    """
    Simulate multiple Poisson processes and return the number of events in each simulation.
    """
    num_events_list = []
    for _ in range(num_simulations):
        event_times, _ = inter_arrival_times(rate, time_limit)
        num_events_list.append(len(event_times))
    return num_events_list


# Parameters
TI =Mean_interarrival_time
rate = 1.0 / TI  # Rate λ
time_limit =1
num_simulations = 10000

num_events_list = simulate_multiple_processes(rate, time_limit, num_simulations)

plt.hist(num_events_list, bins=30, edgecolor='black', alpha=0.7)
plt.title('Histogram of Number of Events per Simulation')
plt.xlabel('Number of Events')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
