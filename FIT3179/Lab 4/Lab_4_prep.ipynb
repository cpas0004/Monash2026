```python
# Decoder lookup array from the lab
decoder = [
    0x00, 0x44, 0x00, 0x54, 0x53, 0x52, 0x48, 0x00, 0x4B,
    0x4C, 0x46, 0x41, 0x00, 0x3F, 0x3E, 0x45, 0x00, 0x01,
    0x02, 0x03, 0x04, 0x05, 0x06, 0x07, 0x08, 0x09, 0x40,
    0x00, 0x49, 0x42, 0x4A, 0x47, 0x51, 0x24, 0x25, 0x26,
    0x27, 0x28, 0x29, 0x2A, 0x2B, 0x2C, 0x2D, 0x2E, 0x2F,
    0x30, 0x31, 0x32, 0x33, 0x34, 0x35, 0x36, 0x37, 0x38,
    0x39, 0x3A, 0x3B, 0x3C, 0x3D, 0x4D, 0x00, 0x4E, 0x43,
    0x00, 0x00, 0x0A, 0x0B, 0x0C, 0x0D, 0x0E, 0x0F, 0x10,
    0x11, 0x12, 0x13, 0x14, 0x15, 0x16, 0x17, 0x18, 0x19,
    0x1A, 0x1B, 0x1C, 0x1D, 0x1E, 0x1F, 0x20, 0x21, 0x22,
    0x23, 0x4F, 0x00, 0x50, 0x00, 0x00
]

def z85_decode_block(chars):
    """
    Takes 5 Z85 characters and returns 4 decoded bytes
    """
    if len(chars) != 5:
        raise ValueError("Must provide exactly 5 characters")

    value = 0

    # Convert to base 85 integer
    for c in chars:
        index = ord(c) - 32 # ASCII offset
        decoded = decoder[index] # lookup value
        value = value * 85 + decoded

    # Extract 4 bytes using shifting + masking
    b0 = (value >> 24) & 0xFF
    b1 = (value >> 16) & 0xFF
    b2 = (value >> 8) & 0xFF
    b3 = value & 0xFF

    return [b0, b1, b2, b3]

encoded1 = "mOcMh"
encoded2 = "gCH:E"
decoded_bytes1 = z85_decode_block(encoded1)
decoded_bytes2 = z85_decode_block(encoded2)


print("Decoded bytes 1:", decoded_bytes1)
print("As ASCII 1:", ''.join(chr(b) for b in decoded_bytes1))
print("Decoded bytes 2:", decoded_bytes2)
print("As ASCII 2:", ''.join(chr(b) for b in decoded_bytes2))
```
